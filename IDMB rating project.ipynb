# code:

```python
srno_list = []
movie_names=[]
director_names=[]
star_name1=[]
star_name2=[]
star_name3=[]
star_name4=[]
duration_list=[]
year_list=[]
metascore_list=[]
rating_list=[]
votes_list=[]
certification=[]


start = 0
end = 34

for page in range(start, end):
    srno_elements = driver.find_elements(By.XPATH, "//span[@class='lister-item-index unbold text-primary']")
   
    for element in srno_elements:
        srno_text = element.text.replace(',', '').replace('.', '')
        srno_list.append(srno_text)
        

   #Movie name 
    movie_elements = driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/a')

    for element in movie_elements:
        movie_name = element.text
        movie_names.append(movie_name)
    
    #director name
    director_elements = driver.find_elements(By.XPATH,'//p[contains(text(), "Director:")]/a[1]')

    for element in director_elements:
        director_name = element.text
        director_names.append(director_name)
        
    stars_elements = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p[3]/a[2]')
    for element in stars_elements:
        stars_name = element.text
        star_name1.append(stars_name)
    
    stars_elements = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p[3]/a[3]')
    for element in stars_elements:
        stars_name = element.text
        star_name2.append(stars_name)
        
    stars_elements = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p[3]/a[4]')

    for element in stars_elements:
        stars_name = element.text
        star_name3.append(stars_name)
    stars_elements = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p[3]/a[5]')

    for element in stars_elements:
        stars_name = element.text
        star_name4.append(stars_name)
    
    duration_element=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
    for element in duration_element:
        duration_text = element.text.replace(',', '').replace('.', '')
        duration_list.append(duration_text)
        
        
   #year
    year=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
    for i in year:
        year_list.append(i.text)
        
    #metascore
    metascore=driver.find_elements(By.XPATH,"//span[@class='metascore  favorable']")
    for i in metascore:
        metascore_list.append(i.text)
        
    #rating_elements
    rating_elements = driver.find_elements(By.XPATH, '//div[@class="ratings-bar"]/div[@class="inline-block ratings-imdb-rating"]/strong')
    for element in rating_elements:
        rating_text = element.text
        rating_list.append(rating_text)    
        
        
    #votes
    votes = driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]')
    for i in votes:
        votes_list.append(i.text[6:].split('|')[0].replace('Votes:',''))
        

    
        
    #certificate
    certificate = driver.find_elements(By.XPATH,'//span[@class="certificate"]')
    for element in certificate:
        certificate_text=element.text
        certification.append(certificate_text)
    
        
    nxt_button = driver.find_element(By.XPATH, '//a[@class="lister-page-next next-page"]')
    next_page_href = nxt_button.get_attribute("href")
    
    try:
        driver.get(next_page_href)
    except:
        # Handle any exceptions that might occur while navigating to the next page
        pass
    
df1=pd.DataFrame(index = list(range(0,1700))) 
df1["SrNO"]=pd.Series(srno_list)
df1["movie_names"]=pd.Series(movie_names)
df1["director_names"]=pd.Series(director_names)
df1["duration_list"]=pd.Series(duration_list)
df1["year_list"]=pd.Series(year_list)
df1["metascore_list"]=pd.Series(metascore_list)
df1["rating_list"]=pd.Series(rating_list)

```

# Gross Collection:

```python
gross_collection=[]
start=0
end=34
for page in range(start,end):
    gross_coll_tags=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
    for i in gross_coll_tags:
        gross_text=i.text.split('|')
        if len(gross_text) >=2:
            gross_collection.append(i.text.split('|')[1].replace('Gross:','').strip())
        else:
            gross_collection.append('-')
            
    nxt_button = driver.find_element(By.XPATH, '//a[@class="lister-page-next next-page"]')
    nxt_button.click()
    time.sleep(3)
print(gross_collection)    
```

# For Genre

```python
import requests
page=requests.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
soup=BeautifulSoup(page.content)

genre_collection = []
start=0
end=34
for page in range(start,end):
    
    for container in soup.find_all("div", class_="lister-item-content"):
        genre_element = container.find("span", class_="genre")
        if genre_element:
            genres = genre_element.get_text(strip=True)
            genre_list = genres.split(",")
            genre_collection.append(genre_list)

    # Separate genre values into three variables
    genre_1 = []
    genre_2 = []
    genre_3 = []

    for genres in genre_collection:
        if len(genres) >= 1:
            genre_1.append(genres[0].strip())
        if len(genres) >= 2:
            genre_2.append(genres[1].strip())
        if len(genres) >= 3:
            genre_3.append(genres[2].strip())

# Print the genre values in three separate variables
print("Genre 1:", genre_1)
print("Genre 2:", genre_2)
print("Genre 3:", genre_3)
```
